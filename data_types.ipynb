#Q4. Explain the difference between the union() and update() methods for a set. Give an example of each method.
union() Method:
The union() method returns a new set containing all the unique elements from the original set as well as the elements from one or more other sets.
It does not modify the original set but instead returns a new set.
update() Method:

The update() method modifies the original set by adding all the elements from one or more other sets.
It updates the set in place and does not return a new set.

```python
# Example of union() method
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1.union(set2)
print(union_set)  
print(set1)       
# Example of update() method
set3 = {5, 6, 7}
set1.update(set3)
print(set1)  
```

Q5. What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.

A dictionary in Python is an unordered collection of key-value pairs. Each key-value pair maps the key to its associated value. Dictionaries are commonly used for storing and retrieving data in a structured way where each piece of data is associated with a unique key.

```python
# Creating a dictionary
my_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

# Accessing elements
print(my_dict["name"])  # Output: John
print(my_dict["age"])   # Output: 30
print(my_dict["city"])  # Output: New York
```

Q7. Using setdefault() method, create key named topics in the given dictionary and also add the value of
the key as this list ['Python', 'Machine Learning’, 'Deep Learning']
dict1 = {'language' : 'Python', 'course': 'Data Science Masters'}

```python

# Given dictionary
dict1 = {'language': 'Python', 'course': 'Data Science Masters'}

# Adding the key 'topics' and assigning the value as the list ['Python', 'Machine Learning', 'Deep Learning']
dict1.setdefault('topics', ['Python', 'Machine Learning', 'Deep Learning'])

# Printing the modified dictionary
print(dict1)
```

Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display
these three view objects for the given dictionary.
dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

In Python, dictionaries provide three view objects that allow you to view the keys, values, and key-value pairs present in the dictionary. These view objects are:

dict_keys: This view object provides a dynamic view of all the keys present in the dictionary.
dict_values: This view object provides a dynamic view of all the values present in the dictionary.
dict_items: This view object provides a dynamic view of all the key-value pairs present in the dictionary.

```python
# Given dictionary
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

# Using the three in-built methods to display the view objects
keys_view = dict1.keys()
values_view = dict1.values()
items_view = dict1.items()

# Displaying the view objects
print("Keys view:", keys_view)
print("Values view:", values_view)
print("Items view:", items_view)
```
